```python
from sympy import *
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
init_printing()
```

```python
x = Symbol('x')
```

```python
n = Symbol ('n', integer=True, positive=True)
```

```python
a=(Integral(-x*sin(n*x),(x,-pi,0))+Integral(x*sin(n*x),(x,0,pi)))/sqrt(pi)
Eq(a,simplify(a.doit()))
```

```python
b=(Integral(-x*cos(n*x),(x,-pi,0))+Integral(x*cos(n*x),(x,0,pi)))/sqrt(pi)
Eq(b,simplify(b.doit()))
```

```python
g=(Integral(-x,(x,-pi,0))+Integral(x,(x,0,pi)))/sqrt(2*pi)
Eq(g,g.doit())
```

```python
c=simplify(b.doit())/sqrt(pi)*cos(n*x)
c
```

```python
d=g.doit()/sqrt(2*pi)+Sum(c,(n,1,10))
Eq(d,d.doit())
```

```python
plot(d.doit(), (x, -pi, pi))
```

```python
d=g.doit()/sqrt(2*pi)+Sum(c,(n,1,30))
```

```python
plot(d.doit(), (x, -pi, pi))
```
